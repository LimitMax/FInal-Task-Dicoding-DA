# Regresi Linear

Analisis regresi adalah metode analisis yang digunakan untuk menganalisis hubungan antara dua atau lebih variabel. Dimana task yang saya tentukan yaitu: Prediksi jumlah sewa sepeda harian berdasarkan lingkungan dan musim dengan multiple regression menggunakan rumus secara manual tanpa bantuan library scikit-learn ataupun tensorflow. Jadi hanya menggunakan rumus.

```python
df2 = df_day.copy()
```

```python
df2.head()
```

```python
df2.corr()
```

```python
# rumus linear regression berganda 𝑦 = 𝑏0 + 𝑏1 * 𝑥1 + 𝑏2 * 𝑥2 + ... + 𝑏𝑛 * 𝑥𝑛

# tentukan y
y = df2['cnt']

# tentukan x
x = df2[['season','temp','hum','windspeed','casual','registered']]

# hitung b1 * x1
b1 = x['season'] * x['temp']

# hitung b2 * x2
b2 = x['season'] * x['hum']

# hitung b3 * x3
b3 = x['season'] * x['windspeed']

# hitung b4 * x4
b4 = x['season'] * x['casual']

# hitung b5 * x5
b5 = x['season'] * x['registered']

# hitung b0
b0 = y - b1 - b2 - b3 - b4 - b5
```

```python
# buat dataframe baru
df3 = pd.DataFrame({'b0':b0, 'b1':b1, 'b2':b2, 'b3':b3, 'b4':b4, 'b5':b5})
```

```python
# hitung y prediksi
y_pred = df3['b0'] + df3['b1'] + df3['b2'] + df3['b3'] + df3['b4'] + df3['b5']
```

```python
# hitung error
error = y - y_pred

# hitung error kuadrat
error2 = error ** 2

# hitung MSE
mse = error2.mean()

# hitung RMSE
rmse = np.sqrt(mse)

# hitung MAE
mae = error.abs().mean()

# hitung MAPE
mape = (error.abs() / y).mean() * 100

# hitung R2
r2 = 1 - (error2.sum() / ((y - y.mean()) ** 2).sum())

# tampilkan hasil
print('MSE:', mse)
print('RMSE:', rmse)
print('MAE:', mae)
print('MAPE:', mape)
print('R2:', r2)
```

```python
# visualisasi hasil
plt.figure(figsize = (8,5))
plt.scatter(x = y, y = y_pred, color = 'red')
plt.plot([y.min(), y.max()], [y.min(), y.max()], color = 'blue', lw = 1, linestyle = '--')
plt.xlabel('Actual')
plt.ylabel('Predict')

plt.tight_layout()
plt.show()
```

**Insight:** Jika dilihat dari perhitungan secara manual, data ini sudah termasuk ke dalam regresi linear. Namun hasil prediksinya masih termasuk overfitting. Karena, nilai MSE yang sangat kecil dan hampir 0 dimana menunjukkan bahwa model terlalu pas dengan data pelatihan. Namun,karena saya menggunakan data dari datasetnya secara langsung,belum menggunakan dataset lain. Maka saya pikir,hasilnya bakal overfit.
